```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
from bauer.utils.data import load_garcia2022
```

```python
df = pd.read_csv('/data/ds-numrisk/derivatives/summary_data/magRiskData2.csv')
df['log(risky/safe)'] = np.log(df['prob_bet'] / df['sure_bet'])
df = df[df['exptype'] != 1]

df['p1'] = 1.0
df['n1'] = df['sure_bet']
df['p2'] = 0.55

df['n2'] = df['prob_bet']


df['format'] = df['exptype'].map({2: 'symbolic', 3: 'non-symbolic'})


df['choice'] = df['choice'].map({-1: True, 1:False, -2: np.nan, 2:np.nan})

df['subject'] = df['subcode'].astype(int)


# Make a 'trial nr' for every format and subject
df['trial_nr'] = df.groupby(['format', 'subject']).cumcount() + 1
```
